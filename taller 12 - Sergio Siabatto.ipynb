```python
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
data = pd.read_csv (r'datos_taller.csv')
data
```

```python
vader = SentimentIntensityAnalyzer()
data['sentimiento_vader'] = data['tweets'].apply(lambda valor: vader.polarity_scores(valor)['compound'])
data.head(20)
```

###  `[25 pts]` Punto 2: ¬øCu√°l es el tuit m√°s positivo y m√°s negativo?

```python
max = data.sentimiento_vader.max()
max
```

```python
positivos = data.loc[data['sentimiento_vader'] == max]
pd.set_option('display.max_colwidth', None)
positivos
```

```python
min = data.sentimiento_vader.min()
min
```

```python
minimos = data.loc[data['sentimiento_vader'] == min]
pd.set_option('display.max_colwidth', None)
minimos
```
